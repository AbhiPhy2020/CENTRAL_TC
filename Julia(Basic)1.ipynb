By convetion function followed by ! sign alter their contents and function lacking ! do not. \

```julia
function pauli_string_to_pauli(ps::String, type_out = UInt64)
    l = zeros(Int64, length(ps))
    for (i, c)=enumerate(ps)
        if c == 'I'
            l[i] = 0
        elseif c == 'X'
            l[i] = 1
        elseif c == 'Y'
            l[i] = 2
        elseif c == 'Z'
            l[i] = 3
        else
            error("Invalid character: $c")
        end
    end
    return pauli_string_to_pauli(l, type_out)
end
```

```julia
using LinearAlgebra

# Define the density matrix ρ
ρ = [0.5 0.3 0.2 0.4;
     0.3 0.2 0.1 0.2;
     0.2 0.1 0.3 0.5;
     0.4 0.2 0.5 0.7]

# Define the dimensions of the subsystems A and B
dim_A = 2
dim_B = 2

# Reshape the density matrix into a matrix of size (dim_A, dim_B, dim_A, dim_B)
ρ_reshaped = reshape(ρ, (dim_A, dim_B, dim_A, dim_B))

# Perform the partial trace over subsystem B
ρ_A = zeros(dim_A, dim_A)
for i in 1:dim_A
    for j in 1:dim_A
        ρ_A[i, j] = sum(ρ_reshaped[i, :, j, :])
    end
end

# Print the resulting density matrix after partial trace
println("Partial Trace (Subsystem A):\n", ρ_A)
```
